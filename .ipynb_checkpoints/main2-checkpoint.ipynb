```python
# imports
from __future__ import print_function
import os
```

```python
scripts_path = "data/friends/scripts/"
scripts = os.listdir(scripts_path)
with open("data/friends/all_scripts.txt",'w') all_scripts:
    for i,script in enumerate(scripts):
        with open(scripts_path+script).read() as script_file:
            all_scripts.write(script_file)
            print(i,script)
```
